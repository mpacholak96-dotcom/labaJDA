# Podstawowe operacje na liczbach
**Ćwiczenie:**

Oblicz wynik wyrażeń:

````python
x = 10
y = 5
print(x + y) # dodawanie
print(x - y) # odejmowanie
print(x * y) # mnożenie
print(x / y) # dzielenie
print(x // y) # dzielenie bez reszty
print(x % y) # reszta z dzielenia
print(x ** y) # potęgowanie
````
